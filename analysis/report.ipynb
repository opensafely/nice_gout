```python
display(HTML("""
<style>

/* ===== Overall style ===== */
.report-style{
  font-family: -apple-system, BlinkMacSystemFont, "Segoe UI", Roboto, Arial,
               "Noto Sans", "Liberation Sans", sans-serif;
  line-height: 1.65;
  color: var(--colour-black);
  max-width: 920px;
  margin: 0 auto;
}
             
/* ===== Colours ===== */
:root{
  --colour-teal:        #228096;
  --colour-teal-75:     #59A0B0;
  --colour-teal-50:     #91C0CB;
  --colour-teal-25:     #C8E0E6;
  --colour-cream:       #DED5CA;
  --colour-cream-75:    #E6E0D7;
  --colour-cream-50:    #EEEAE4;
  --colour-cream-25:    #F7F4F1;
  --colour-blue:        #00436C;
  --colour-blue-75:     #407291;
  --colour-blue-50:     #80A1B5;
  --colour-blue-25:     #BFD0DA;
  --colour-green:       #37906D;
  --colour-tan:         #D07B4D;
  --colour-yellow:      #EAD054;
  --colour-pink:        #EDD8CD;
  --colour-black:       #0e0e0e;
  --colour-white:       #ffffff;
}             

/* ===== Report header / banner ===== */
.report-header{
  background: var(--colour-blue);
  color: var(--colour-white);
  padding: 2rem 2.25rem 1.75rem;
  border-radius: 4px;
  margin: 0 0 2rem;
  position: relative;
  overflow: hidden;
}
.report-header::after{
  content: "";
  position: absolute;
  top: 0; right: 0;
  width: 180px; height: 100%;
  background: linear-gradient(135deg, transparent 40%, var(--colour-teal) 100%);
  opacity: 0.35;
}
.report-header h1{
  font-size: 1.65rem;
  font-weight: 700;
  margin: 0 0 0.3rem;
  letter-spacing: -0.01em;
  position: relative;
  z-index: 1;
}
.report-header .subtitle{
  font-size: 1.3rem;
  opacity: 0.88;
  margin: 0;
  font-weight: 500;
  position: relative;
  z-index: 1;
}
.report-header .header-rule{
  display: block;
  width: 64px;
  height: 3px;
  background: var(--colour-teal-50);
  border: none;
  margin: 0.9rem 0 0.7rem;
  position: relative;
  z-index: 1;
}

/* ===== Section headings ===== */
.report-style h2{
  font-size: 1.45rem;
  font-weight: 700;
  color: var(--colour-blue);
  margin: 2rem 0 0.5rem;
  padding-bottom: 0.35rem;
  border-bottom: 3px solid var(--colour-teal);
}
.report-style h3{
  font-size: 1.3rem;
  margin: 1.5rem 0 0.4rem;
  font-weight: 700;
  color: var(--colour-blue);
}
.report-style h4{
  font-size: 1.2rem;
  margin: 1rem 0 0.5rem;
  font-weight: 700;
  color: var(--colour-teal);
}

/* ===== Body text and links ===== */
.report-style p{
  font-size: 1rem;
  margin: 0 0 0.75rem;
}
.report-style a{
  color: var(--colour-teal);
  text-decoration: underline;
  text-underline-offset: 2px;
}
.report-style a:hover{
  color: var(--colour-blue);
}

/* ===== Recommendation call-out boxes ===== */
.report-rec{
  margin: 0.75rem 0 1.25rem;
  padding: 0.15rem 0;
}
.report-rec-box{
  background: var(--colour-cream-25);
  border-left: 4px solid var(--colour-teal);
  padding: 1rem 1.25rem;
  margin: 0.75rem 0 1.25rem;
  border-radius: 0 4px 4px 0;
}
.report-rec-box .rec-num{
  display: inline-block;
  font-size: 0.8rem;
  font-weight: 700;
  text-transform: uppercase;
  letter-spacing: 0.04em;
  color: var(--colour-white);
  background: var(--colour-teal);
  padding: 0.15rem 0.6rem;
  border-radius: 3px;
  margin: 0 0 0.6rem;
}
.report-rec-box p{
  margin: 0;
  font-size: 0.97rem;
}
.report-rec-box em{
  font-style: italic;
  color: #333;
}
.report-rec-box ul{
  margin: 0.4rem 0 0.2rem 1.2rem;
  padding: 0;
  font-size: 0.97rem;
  color: #333;
  font-style: italic;
}
.report-rec-box ul li{
  margin-bottom: 0.2rem;
}

/* ===== Description text ===== */
.report-desc{
  margin: 0.5rem 0 1rem;
  padding: 0;
  font-size: 0.97rem;
  color: #444;
}
.report-desc p{
  margin: 0;
}
                          
/* ===== Figure / image containers ===== */
.report-figure{
  width: 100%;
  margin: 1rem 0 1.5rem;
  padding: 0;
  background: none;
  border: none;
  border-radius: 0;
  box-sizing: border-box;
}
.report-figure svg,
.report-figure img{
  display: block;
  margin: 0 auto;
  width: 100%;
  max-width: 100%;
  height: auto;
}

/* ===== Collapsible sections ===== */
details.report-collapsible{
  margin: 1.5rem 0;
  border: none;
  border-radius: 0;
  overflow: visible;
}

details.report-collapsible > summary{
  display: flex;
  align-items: center;
  gap: 0.5rem;
  cursor: pointer;

  padding: 0.85rem 1.25rem;
  font-weight: 700;
  font-size: 1.3rem;
  color: var(--colour-blue);

  background: var(--colour-cream-25);
  border: 1px solid var(--colour-cream);
  border-radius: 12px;

  list-style: none;
  user-select: none;
  transition: background 0.15s;
}

details.report-collapsible > summary::-webkit-details-marker,
details.report-sub-collapsible > summary::-webkit-details-marker{
  display: none;
}

/* standard-style triangle (font-independent) */
details.report-collapsible > summary::before{
  content: "";
  width: 0;
  height: 0;
  border-style: solid;
  border-width: 5px 0 5px 7px;  /* native-like proportions */
  border-color: transparent transparent transparent var(--colour-teal);
  margin-right: 0.45rem;
  transition: transform 0.2s ease;
  flex-shrink: 0;
}
             
details.report-collapsible[open] > summary::before{
  transform: rotate(90deg);
}

details.report-collapsible > summary:hover{
  background: var(--colour-cream-50);
}

/* body padding for top-level collapsible */
details.report-collapsible > .collapsible-body{
  padding: 1rem 1.25rem 0.5rem;
}

/* ===== Nested / sub-collapsible ===== */
details.report-sub-collapsible{
  margin: 1rem 0;
  border-left: 3px solid var(--colour-teal-50);
  padding-left: 0;
}

details.report-sub-collapsible > summary{
  display: flex;
  align-items: center;
  cursor: pointer;

  padding: 0.5rem 0.85rem;
  font-weight: 700;
  font-size: 1.05rem;
  color: var(--colour-teal);

  background: var(--colour-cream-25);
  border: 1px solid var(--colour-cream);
  border-radius: 10px;

  list-style: none;
  user-select: none;
  transition: background 0.15s, border-color 0.15s, color 0.15s;
}

/* nested triangle */
details.report-sub-collapsible > summary::before{
  content: "";
  width: 0;
  height: 0;
  border-style: solid;
  border-width: 4px 0 4px 6px;
  border-color: transparent transparent transparent var(--colour-teal-50);
  margin-right: 0.35rem;
  transition: transform 0.2s ease;
  flex-shrink: 0;
}
             
details.report-sub-collapsible[open] > summary{
  background: var(--colour-teal-25);
  border-color: var(--colour-teal-50);
  color: var(--colour-blue);
}

details.report-sub-collapsible[open] > summary::before{
  transform: rotate(90deg);
}

details.report-sub-collapsible > summary:hover{
  color: var(--colour-blue);
}

/* body padding for nested collapsible */
details.report-sub-collapsible > .sub-body{
  padding: 0.5rem 0.75rem 0.25rem;
}
             
/* ===== Smaller subsection collapsibles ===== */
.report-small{
  font-size: 0.9rem;
}

.report-small details.report-collapsible > summary{
  font-size: 1.05rem;
  padding: 0.65rem 1.0rem;
}

.report-small .collapsible-body{
  padding: 0.75rem 1.0rem 0.35rem;
}

.report-small .report-desc,
.report-small .report-methods{
  font-size: 0.9rem;
  color: #444;
}             

/* ===== Section divider ===== */
.report-divider{
  border: none;
  border-top: 2px solid var(--colour-cream);
  margin: 2rem 0;
}

/* ===== Introduction paragraph styling ===== */
.report-intro{
  background: var(--colour-cream-25);
  border-radius: 4px;
  padding: 1.25rem 1.5rem;
  margin: 0 0 1.5rem;
  font-size: 0.97rem;
  line-height: 1.7;
}
.report-intro p{
  margin: 0 0 0.85rem;
}
.report-intro p:last-child{
  margin-bottom: 0;
}

/* ===== Methods box ===== */
.report-methods{
  background: var(--colour-cream-25);
  border-left: 3px solid var(--colour-cream);
  padding: 0.75rem 1rem;
  font-size: 0.93rem;
  color: #555;
  border-radius: 0 4px 4px 0;
}
.report-methods p{
  margin: 0;
}
             
/* ===== Tabs ===== */
.report-tabs-figure { margin-top: 2rem; }

.report-tabs{
  display: flex;
  flex-wrap: wrap;
  gap: 0.4rem;
  margin: 0 0 0.6rem;
}

.report-tab{
  appearance: none;
  border: 1px solid var(--colour-cream);
  background: var(--colour-cream-25);
  color: var(--colour-blue);
  font-weight: 700;
  font-size: 0.95rem;
  padding: 0.35rem 0.65rem;
  border-radius: 999px;
  cursor: pointer;
  transition: background 0.15s, border-color 0.15s, transform 0.05s;
}

.report-tab:hover{ background: var(--colour-cream-50); }
.report-tab:active{ transform: translateY(1px); }

.report-tab.active{
  background: var(--colour-teal-25);
  border-color: var(--colour-teal-50);
  color: var(--colour-blue);
}

.report-tabs-panel{
  width: 100%;
}             

</style>
"""))

base_tabs = [
    ("Overall", ""),
    ("Sex", "_sex"),
    ("Age", "_agegroup"),
    ("Ethnicity", "_ethnicity"),
    ("Socioeconomic status", "_imd"),
    ("Region", "_region"),
    ("Time series", "_itsa")
]

def safe_read_svg(path):
    """
    Read an SVG file and return its contents.

    Returns an empty string if the file does not exist
    or cannot be read.
    """
    path = Path(path)

    if not path.is_file():
        return ""

    try:
        return path.read_text(encoding="utf-8")
    except (OSError, UnicodeDecodeError):
        return ""


def make_unique_id(value):
    """
    Create a stable HTML-safe identifier from a value.
    """
    return hashlib.md5(
        str(value).encode("utf-8")
    ).hexdigest()[:12]


def build_tab_buttons(svg_stem, include_itsa=True):
    """
    Build tab buttons and embedded SVG panels.

    Missing SVG files are omitted. If no SVG files exist,
    empty strings are returned.
    """
    tabs = (
        base_tabs
        if include_itsa
        else [
            tab
            for tab in base_tabs
            if tab[1] != "_itsa"
        ]
    )

    available_tabs = []

    for label, suffix in tabs:
        svg_path = Path(f"{svg_stem}{suffix}.svg")
        svg_content = safe_read_svg(svg_path)

        if svg_content:
            available_tabs.append(
                {
                    "label": label,
                    "suffix": suffix,
                    "content": svg_content,
                }
            )

    # No figures available
    if not available_tabs:
        return "", ""

    figure_id = make_unique_id(svg_stem)

    buttons_html = ""
    panels_html = ""

    for i, tab in enumerate(available_tabs):
        active = i == 0
        active_class = "active" if active else ""
        aria_selected = "true" if active else "false"
        display_style = "block" if active else "none"

        panel_id = f"figure-{figure_id}-panel-{i}"
        button_id = f"figure-{figure_id}-tab-{i}"

        buttons_html += f"""
        <button type="button"
                id="{button_id}"
                class="report-tab {active_class}"
                role="tab"
                aria-selected="{aria_selected}"
                aria-controls="{panel_id}"
                data-panel-id="{panel_id}"
                onclick="
                    const figure = this.closest(
                        '.report-tabs-figure'
                    );

                    figure.querySelectorAll(
                        '.report-tabs-panel'
                    ).forEach(panel => {{
                        panel.style.display = 'none';
                    }});

                    const selectedPanel = figure.querySelector(
                        '#' + this.dataset.panelId
                    );

                    if (selectedPanel) {{
                        selectedPanel.style.display = 'block';
                    }}

                    figure.querySelectorAll(
                        '.report-tab'
                    ).forEach(button => {{
                        const isActive = button === this;

                        button.classList.toggle(
                            'active',
                            isActive
                        );

                        button.setAttribute(
                            'aria-selected',
                            isActive ? 'true' : 'false'
                        );
                    }});
                ">
            {escape(str(tab["label"]))}
        </button>
        """

        panels_html += f"""
        <div id="{panel_id}"
             class="report-tabs-panel"
             role="tabpanel"
             aria-labelledby="{button_id}"
             style="display:{display_style}; width:100%;">
            {tab["content"]}
        </div>
        """

    return buttons_html, panels_html


def tabbed_svg_figure(
    svg_stem,
    max_width=1200,
    aria_label="Figure tabs",
    include_itsa=True
):
    """
    Return a tabbed figure with SVGs embedded directly in the HTML.

    Missing individual SVGs are omitted. If no SVGs exist,
    a blank figure container is returned.
    """
    buttons_html, panels_html = build_tab_buttons(
        svg_stem,
        include_itsa=include_itsa
    )

    # Return a blank area if no files exist
    if not panels_html:
        return f"""
        <div class="report-figure"
             style="
                 max-width:{max_width}px;
                 width:100%;
             ">
        </div>
        """

    return f"""
    <div class="report-figure report-tabs-figure"
         style="
             max-width:{max_width}px;
             width:100%;
         ">

        <div class="report-tabs"
             role="tablist"
             aria-label="{escape(str(aria_label))}">
            {buttons_html}
        </div>

        <div class="report-tabs-panels"
             style="width:100%;">
            {panels_html}
        </div>

    </div>
    """

def collapsible_image(title, path, description=None, max_width=1200, open_by_default=False):
    path = Path(path)
    suffix = path.suffix.lower()
    open_attr = "open" if open_by_default else ""

    if suffix == ".svg":
        content = safe_read_svg(path)
    else:
        content = f'<img src="{path}" />' if path.is_file() else ""

    desc_html = f'<div class="report-desc"><p>{description}</p></div>' if description else ""

    display(HTML(f"""
    <details class="report-collapsible" {open_attr}>
      <summary>{title}</summary>
      <div class="collapsible-body">
        {desc_html}
        <div class="report-figure" style="max-width:{max_width}px;">
          {content}
        </div>
      </div>
    </details>
    """))


def collapsible_image_html(title, path, description=None, max_width=1200, open_by_default=False):
    svg = safe_read_svg(path)

    open_attr = " open" if open_by_default else ""

    desc_html = ""
    if description is not None:
        desc_text = description.strip().replace("\n", " ")
        desc_html = f'<div class="report-desc"><p>{desc_text}</p></div>'

    return f"""
    <details class="report-collapsible" {open_attr}>
      <summary>{title}</summary>
      <div class="collapsible-body">
        {desc_html}
        <div class="report-figure" style="max-width:{max_width}px;">
          {svg}
        </div>
      </div>
    </details>
    """

```
